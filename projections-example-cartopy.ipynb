```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt

def main():
    # Create a list of integers for the utm zones from 1 to 46
    # Would be great to display all the 60 zones, but cartopy is resource intensive
    # continuously encountered memory issues on my computer
    zones = range(1, 46, 1)

    # Create a figure with a black face color
    fig = plt.figure(figsize=(12, 6), facecolor='black')

    # Loop through each zone in the list
    for i, zone in enumerate(zones):
        # Add GeoAxes object with specific UTM zone projection to the figure
        ax = fig.add_subplot(1, len(zones), i + 1,
                             projection=ccrs.UTM(zone=zone,
                                                 southern_hemisphere=True))

        # Set the background color of the subplot to black
        ax.set_facecolor('black')

        # Add coastlines, gridlines, and zone number for the subplot
        ax.stock_img()
        ax.set_title(zone, fontsize=8, color='white')

        # Display the progress
        print(f"Processing image {i+1} of {len(zones)}")

    # Add a supertitle for the figure
    fig.suptitle("UTM Projection - Zones", color='white')

    # Adjust the spacing between subplots
    plt.subplots_adjust(wspace=0.001)  # Decrease the horizontal spacing between subplots

    plt.show()

if __name__ == '__main__':
    main()
```

```python
# This example showcases how to display various projection zones.
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

# List of projections to display
projections = [
    ccrs.PlateCarree(),
    ccrs.Mercator(),
    ccrs.LambertCylindrical(),
    ccrs.Robinson(),
    ccrs.Orthographic(),
    ccrs.InterruptedGoodeHomolosine(),
    ccrs.Mollweide(),
    ccrs.EquidistantConic(),
    ccrs.EckertIV(),
    ccrs.NearsidePerspective(),
    ccrs.Sinusoidal(),
    ccrs.SouthPolarStereo(),
    ccrs.NorthPolarStereo(),
    ccrs.Stereographic(),
    ccrs.AlbersEqualArea(),
    ccrs.AzimuthalEquidistant(),
    ccrs.Gnomonic(),
    ccrs.Geostationary(),
    ccrs.TransverseMercator(),
    ccrs.OSNI(),
    ccrs.Miller(),
    ccrs.PlateCarree(central_longitude=180),
    ccrs.Geostationary(),
    ccrs.Geocentric(),
    ccrs.RotatedPole(),
]

# Set up the figure and axes
fig, axs = plt.subplots(6, 3, figsize=(15, 15))

# Iterate through the list of projections
for i, ax in enumerate(axs.flat):
    if i < len(projections):
        proj = projections[i]

        # Create a new GeoAxes with the current projection
        ax = plt.subplot(6, 3, i+1, projection=proj)

        # Display the stock image
        ax.stock_img() # Instead of stockimag, you can also employ the vector data

        # Get the projection name
        proj_name = proj.__class__.__name__

        # Annotate the plot with the projection name
        ax.text(0.5, 1.05, proj_name, transform=ax.transAxes,
                horizontalalignment='center', verticalalignment='bottom',
                fontsize=12, color='black')


# Adjust the layout and spacing of subplots
fig.tight_layout()

# Show the plot
plt.show()
```
